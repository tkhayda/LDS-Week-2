```python
from py2neo import Node
for i in range(len(df1)):
    test = Node("Person", member_ID = str(df1['member_ID'][i]), first_name = df1['member_fname'][i], last_name=df1['member_lname'][i], year_joined=str(df1['year_joined'][i]), semester_hours=str(df1['semester_hours'][i]), exec_id=str(df1['exec_id'][i]), faction_id=str(df1['faction_id'][i]), favrep_id=str(df1['favrep_id'][i]))
    graph.create(test)
```

```python
for i in range(len(df2)):
    test = Node("Role", exec_id=str(df2['exec_id'][i]), position=df2['position'][i], elected=df2['elected'][i])
    graph.create(test)
```

```python
for i in range(len(df3)):
    test = Node("Faction", faction_id=str(df3['faction_id'][i]), species=df3['species'][i], mascot_name=df3['mascot_name'][i], color=df3['Color'][i])
    graph.create(test)
```

```python
for i in range(len(df4)):
    test = Node("FavoriteReptiles", faverep_id=str(df4['faverep_id'][i]),reptile=df4['reptile'][i], lifespan_years=str(df4['lifespan_years'][i]), diet=df4['diet'][i], conservation_status=df4['conservation_status'][i])
    graph.create(test)
```

```python
from py2neo import Relationship
faction = Relationship.type("IS_PART_OF")
role = Relationship.type("SERVES_AS")
rep=Relationship.type("LIKES")
```

```python
graph.run("MATCH (a:Person),(b: Faction) WHERE a.faction_id=b.faction_id CREATE (a)-[:IS_PART_OF]->(b);")
graph.run("MATCH (a:Person),(b: Role) WHERE a.exec_id=b.exec_id CREATE (a)-[:SERVES_AS]->(b);")
graph.run("MATCH (a:Person),(b: FavoriteReptiles) WHERE a.favrep_id=b.faverep_id CREATE (a)-[:LIKES]->(b);")
```
